Factoiral

```python
def fact_number(n):
    fact = 1
    for i in range (1, n+1):
        fact = fact *i
    return fact
fact_number(5)
```

```python
# sum of all natural numbers till the number provided

def sum_numbers(n):
    num = 0

    for i in range (1, n+1):
        num = num+i

    return num
sum_numbers(10)
```

Lambda Functions

```python
# Create a lambda function that adds 10 to the number passed as an arguement, and print the result

x = lambda a :a+45

print(x(6))
```

```python
# Create a lamda function that checks a number is even or odd

x = lambda a: "even" if a%2==0 else "odd"

print(x(51))
```

```python
print(x(4))
```

```python
# Create a lambda function that adds two integers

x = lambda a, b: a+b
print(x(5,6))
```
