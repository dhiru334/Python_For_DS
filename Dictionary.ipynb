Dictionary - Dictionary stores element as keys and values pairs.
The key like an index, its always unique and immutable. 
The values are the objects that contain information. 
Values are accessed using their keys.
Each key is followed by a value separated by a colon.
The values can be immutable, mutable and duplicates.
Each Key and value pair is separated by a comma enclosed inside a curly brackets.


```python
#Creating a dictionary


d1 = {"India": "INR", "USA": "USD", "Hong Kong": "HKD"}
# Accessing values using keys

d1["India"]
```

```python
# Replacing the values for a key in a dictionary

d1["India"] = "NEW"

d1
```

```python
# Inserting a new key-values pair

d1["Japan"] = "YEN"

d1
```

```python
# Deleting a key value pair

del d1["Japan"]

d1
```

```python
# Sorting a dictionary

sorted(d1)
```
