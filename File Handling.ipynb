#### File Handling


```python
# Readig a file

file = open('txtx1.txt', 'r')

for line in file:
    print(line)
```

```python
file = open('txtx1.txt', 'r')

print(file.read())
```

```python
file = open('txtx1.txt', 'r')
print(file.read(5))
```

```python
file = open('txtx1.txt', 'r')
print(file.readline())
```

```python
# Writing a file

file = open('txtx1.txt', 'w')
file.write("This is a write operation")
file.write("\nHello!, Bangaluru")
file.close()
```

```python
file = open('txtx1.txt', 'r')
print(file.read())
```

```python
# with operation

with open('txtx1.txt', 'a') as file:
    file.write("\nImpact")
```

```python
file = open('txtx1.txt', 'r')
for line in file:
    print(line)
```
