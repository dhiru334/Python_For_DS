```python
#List Concat

list1 = [1,2,3,4]
list2 = [4,5,6,7,8,9]

new_list = list1 + list2
print(new_list)
len(new_list)
```

```python
#Membership in lists & tuples

tuple1 = (1,2,3,4,5,6)
print(1 in tuple1)
print(7 in tuple1)
```

Mutability - means can be change

```python
#extend() function - it extends the list

list1 = [1,2,3,4]
list1.extend([6, 7])


print(len(list1))

list1
```

```python
#append() function 

list1 = [3,4,5,6]
list1.append([7, 8, 9])

list1
```

```python
#delete command 

list1 = [1,3, 4, 5,6]

del list1[3]
list1
```

```python
#pop() - Deletes the object based on the index value
#Default - Removes the last element 

list1 = [10, 20, 30, 40, 50]
list1.pop(3)
list1
```

```python
list1 = [11, 22, 33, 44]
list1.remove(22)
list1
```
