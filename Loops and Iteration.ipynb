Iteration

```python
#Iteration over a list of integers

list_1 = [10, 20, 30, 40, 50]
for num in list_1:
    print(num)
```

```python
# Iterate over a string

str_1 = "Bangaluru"
for ch in str_1:
    print(ch)
```

```python
# Iterate over a dictionary


students_data = {1:["Dheeraj", 21, "DS"], 2:["Suraj", 23, "Softwar"], 3:["Jadhav", 33,"HR"]}

for key, val in students_data.items():
    print(val)
                                             
```

```python
for key in students_data.keys():
    print(key)
```

```python
# Iterate over a range of values

for i in range(1, 8):
    print(i)
```

Comprehensions

```python
list_1 = ["Automobiles", "Honda", "Benz", "Maruti", "Kia"]

list_2 = []

for i in list_1:
    list_2.append(i)
print(list_2)
```

```python
list_1 = ["Automobiles", "Honda", "Benz", "Maruti", "Kia"]

list_2 =[]

for i in list_1:
    list_2.append(len(i))
print(list_2)
```

```python
list_1 = ["Automobiles", "Honda", "Benz", "Maruti", "Kia"]

list_2 = [len(i) for i in list_1]
print(list_2)
```

```python
list_1 = ["Automobiles", "Honda", "Benz", "Maruti", "Kia"]

d = {i: len(i) for i in list_1}

d
```
