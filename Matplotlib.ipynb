```python
plt.bar(churn_df['Geography'], churn_df['EstimatedSalary'].mean(), color='cyan', edgecolor = 'orange')
plt.title('Geography vs EstimatedSalary')
plt.show()
```

```python
plt.bar(churn_df['Geography'], churn_df['EstimatedSalary'].max(), color='cyan', edgecolor='orange')
plt.title('Geography vs Esimated Salary', fontdict  = {'fontsize':20, 'color':'Green'})
plt.show()
```

```python
plt.bar(churn_df['Gender'], churn_df['Age'].max(), color='orange', edgecolor='black')
plt.title('Gender vs Age', fontdict = {'fontsize':20, 'color':'Green'})
plt.show()
```

```python
plt.scatter(churn_df['Age'], churn_df['Tenure'])
plt.title('Scatter Plot')
plt.xlabel('This is X Axis')
plt.ylabel('This is Y Axis')
plt.show()
```

```python
plt.hist(churn_df['Tenure'], bins=30)
plt.ylabel('Probability')
plt.xlabel('Tenure')
plt.show()
```

```python
labels_ex = 'javaScript', 'Java', 'Python', 'R'
sizes = [15,23,30,44]
explode_labels = (0, 0.5, 0, 0)
fig1, ax1 = plt.subplots()

ax1.pie(sizes, explode = explode_labels, labels= labels_ex, shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

Multiple Plots on Same Graphe

```python
x = [1,2,3,4,5]
y1= [10,20,30,40,50]
y2 = [15,25,35,45,55]

plt.plot(x,y1, label='Product A')
plt.plot(x,y2, label='Product B')
plt.show()
```

Subplots (Multiple Graphs)


```python
plt.subplot(1,2,1)
plt.plot([1,2,3],[3,2,1])
plt.title('Graph 1')

plt.subplot(1,2,2)
plt.plot([1,2,3],[3,2,1])
plt.title('Graph 2')

plt.show()

plt.savefig('two_graphs.png', dpi=300,bbox_inches='tight')
plt.show()
```
