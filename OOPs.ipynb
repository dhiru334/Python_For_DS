## Parametrised Constructor

parametrised constructor - dynamically assign the attribute values during object creation

```python
class rectangle :
    def __init__(self):
        self.length = 10
        self.breadth = 15
```

```python
class rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
```

```python
# Instantiating a class
rect = rectangle(5, 18)

# Calling the instance variable
rect.breadth
```

# Adding a method to class

calculate_area() - returns the product of attributes length and breadth 

self  - identifies its association with the instance

```python
class Rectangle :
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def calculate_area(self):
        return self.length * self.breadth
```

```python
# Instantiating the rectangle class
rect = Rectangle(10, 5)
```

```python
rect.length
```

```python
rect.breadth
```

```python
rect.calculate_area()
```

## Inheritance and Overriding

```python
class Employee:
    def function(self):
        print("hello world")

class Department(Employee):
    pass
```

```python
emp = Employee()
emp.function()
```

```python
dept = Department()
dept.function()
```

```python
class shape:
    def set_color(self, color):
        self.color = color

    def calculate_area(self):
        pass

    def color_the_Shape(self):
        color_price = {"red" : 10, "blue" : 5, "green" : 9}
        return self.calculate_area()* color_price[self.color]
```

```python
class Circle(shape):
    pi = 3.14
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return Circle.pi*self.radius
```

```python
c = Circle(5)
# Accessing set_colot from parent class, because of inheritance

c.set_color("red")
print("Circle with radius =", c.radius, "when colored", c.color, "costs $", c.color_the_Shape())
```

```python
c.calculate_area()
```

```python
class Rectangle(shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    # Overriding user defined method
    def calcuclate_area(self):
        return self.length * self.breadth

    # Overriding python default method
    def __str__(self):
        return "area of rectangle = " + str(self.calculate_area())
```

```python
r = Rectangle(5, 10)
r.set_color("blue")


print("Rectangle with length =", r.length," and breadth = ", r.breadth, "when colored", r.color, "costs $", r.color_the_shape()) 
```
